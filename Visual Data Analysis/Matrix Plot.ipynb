```python
#impporting the modules
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
#generating 2-D 10*10 matrix of random number from 1 to 100
data=np.random.randint(low=1,high=100,size=(10,10))
#setting the annot
annot=True
#plotting the heatmap
sns.heatmap(data,annot=annot)
```

```python
#generating 2-D 10*10 matrix of random number from 1 to 100
data=np.random.randint(low=1,high=100,size=(10,10))
#setting the annot
annot=True
#plotting the heatmap
sns.heatmap(data,annot=annot,cmap="tab20")#changing the colour theme
```

```python
#generating 2-D 10*10 matrix of random number from 1 to 100
data=np.random.randint(low=1,high=100,size=(10,10))
#setting the annot
annot=True
#plotting the heatmap
sns.heatmap(data,annot=annot,linewidths=2,linecolor="yellow")#changing the linecolur and linewidth
```

```python
#generating 2-D 10*10 matrix of random number from 1 to 100
data=np.random.randint(low=1,high=100,size=(10,10))
#setting the annot
annot=True
#plotting the heatmap
sns.heatmap(data,annot=annot,cbar=False)#setting the colourbar to false 
```

```python
data=np.random.randint(low=1,high=100,size=(10,10))
#setting the annot
annot=True
#plotting the heatmap
sns.heatmap(data,annot=annot,xticklabels=False,yticklabels=False)#without x and y labels
```

```python
flights=sns.load_dataset("flights")
flights
```

```python
a
```

```python
pvf=flights.pivot_table(values="passengers",index="month",columns="year")# Pivot table of passengers with months as rows and years as columns
```

```python
sns.heatmap(pvf)
```

```python
sns.heatmap(pvf,cmap="magma",linecolor="white",linewidths=1)# Heatmap of passenger pivot table using magma colormap with white grid lines
```
