```python
import matplotlib.pyplot as plt
import numpy as np
fig,axes=plt.subplots(1,3,figsize=(12,4))
x=np.arange(1,11)
axes[0].plot(x,x**3,"g",lw=2)
axes[0].grid(True)
axes[0].set_title("deafault Grid")
axes[1].plot(x,np.exp(x),"r")
axes[1].grid(color="b",ls="--",lw=0.25)
axes[1].set_title("custom Grid")
axes[2].plot(x,x)
axes[2].set_title("NO Grid")
fig.tight_layout()
plt.show()
# Plot with three subplots comparing default grid, custom grid, and no grid styles
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.grid(True)
ax.spines["bottom"].set_color("blue")
ax.spines["left"].set_color("red")
ax.spines["left"].set_linewidth(2)
ax.spines["right"].set_color("black")
ax.spines["top"].set_color("green")
ax.plot([5,6,7,8,1])
plt.show()

# Plot with customized axis spines (colored and styled) and grid enabled



```

```python
#using iris datasets
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris=sns.load_dataset("iris")
iris.head()
```

```python
#pairplot
sns.pairplot(data=iris)
```

```python
sns.pairplot(iris,hue="species",palette="rainbow")#changing the colour
```

```python
sns.PairGrid(data=iris)#empty canvas
```

```python
#mapping the data
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
#3 different type of plotting in the pairplot
```

```python
#using tips dataset
tips=sns.load_dataset("tips")
tips.head()
```

```python
g=sns.FacetGrid(tips,col="time",row="smoker")
g.map(plt.hist,"total_bill")
```

```python
g=sns.FacetGrid(tips,col="time",row="smoker",hue="sex",palette="coolwarm")
g.map(plt.scatter,"total_bill","tip").add_legend()
#facegrid with hue and different palette
```

```python
g=sns.JointGrid(x="total_bill",y="tip",data=tips)
#empty canvas
```

```python
g=g.plot(sns.regplot,sns.displot)
```

```python
#implemenatation of matplotlib fucntion
import matplotlib.pyplot as plt
import numpy as np
#dummy data
x1=np.linspace(0.0,5.0)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y1
```

```python
x1
```

```python
#creating 2 subplots
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(12,5))
#plot without grid
ax1.plot(x1,y1)
ax1.set_title("Plot without grid")
#plot with grid
ax2.plot(x1,y1)
ax2.set_title("Plot with grid")
#drawing gridlines
ax2.grid(True)
plt.show()
```

```python
#implemenatation of matplotlib fucntion
import matplotlib.pyplot as plt
import numpy as np
#dummy data
x1=np.linspace(0,2*np.pi,400)
y1=np.sin(x1**2)
#setting graph color
plt.plot(x1,y1,"green")
#setting title
plt.title("Plotting with linewidth and linestyle")
#drawing gridlines of grey color
plt.grid(True,color="grey",linewidth=0.4,linestyle=":")
plt.show()
```
