```python
import seaborn as sns
sns.set_theme()
tips=sns.load_dataset("tips")#loading the tips dataset 
```

```python
tips
```

```python
sns.barplot(x="sex",y="total_bill",data=tips)#bar plot to plot the total_bill gender wise
```

```python
import numpy as np
```

```python
sns.countplot(x="day",data=tips)#countplot for 4 days
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,palette="rainbow")
# Boxplot of total_bill grouped by day using a rainbow color palette
```

```python
sns.boxplot(data=tips,palette="rainbow",orient="h")# Horizontal boxplots of all numerical columns in tips dataset with rainbow palette
```

```python
sns.boxplot(x="day",y="total_bill",hue="sex",data=tips,palette="rainbow")# Boxplot of total_bill by day with separate boxes for sex using rainbow palette
```

```python
sns.boxplot(x="day",y="total_bill",hue="smoker",data=tips,palette="rainbow")# Boxplot of total_bill by day with separate boxes for smoker status using rainbow palette
```

```python
sns.violinplot(x="day",y="total_bill",data=tips,palette="rainbow")# Violin plot of total_bill distribution by day using rainbow palette
```

```python
sns.violinplot(x="day",y="total_bill",hue="sex",data=tips,palette="rainbow")# Violin plot of total_bill by day split by sex using rainbow palette
```

```python
sns.displot(data=tips,x="total_bill",col="time",kde=True)# Distribution plots of total_bill for each time category with KDE overlay
```

```python
sns.distplot(tips["total_bill"])# Histogram and KDE plot of total_bill distribution
```

```python
sns.distplot(tips["total_bill"],kde=True,bins=5)# Histogram of total_bill with 5 bins and KDE curve
```

```python
sns.distplot(tips["total_bill"],kde=True,bins=5*6)# Histogram of total_bill with 30 bins and KDE curve
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="scatter")# Scatter plot joint distribution of total_bill vs tip with marginal histograms
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="hex",color="red")# Hexbin joint plot of total_bill vs tip with red color mapping
```

```python
sns.jointplot(x="total_bill",y="tip",data=tips,kind="reg",color="red")# Regression joint plot of total_bill vs tip with red regression line
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue="sex",palette="coolwarm")
```

```python
sns.rugplot(tips['day'])
```

```python
sns.kdeplot(tips["tip"])
sns.rugplot(tips["tip"])# KDE plot of tip distribution with rug plot showing individual data points
```

```python
sns.kdeplot(tips["total_bill"])
sns.rugplot(tips["total_bill"])# KDE plot of total_bill distribution with rug plot for individual observations
```
