```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset("tips")#loading the tips dataset
```

```python
tips
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips)# Scatter plot of total_bill vs tip with fitted linear regression line
```

```python
sns.set_style("whitegrid")
sns.lmplot(x="total_bill",y="tip",data=tips)# Scatter plot with regression line on a whitegrid-style background
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="time")# Scatter plot with regression lines of total_bill vs tip separated by time of day
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"],palette="coolwarm")# Scatter plot with regression lines of total_bill vs tip separated by sex using custom markers and coolwarm palette
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="time",markers=["o","v"],palette="coolwarm")# Scatter plot with regression lines of total_bill vs tip separated by time using custom markers and coolwarm palette
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"],palette="plasma",scatter_kws={"s":100})# Regression plot of total_bill vs tip by sex with large markers and plasma palette
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"],palette="plasma",scatter_kws={"s":100*7.8})# Regression plot of total_bill vs tip by sex with enlarged markers using plasma palette
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="sex",markers=["o","v"],palette="plasma",scatter_kws={"s":50},line_kws={"lw":10})# Regression plot of total_bill vs tip by sex with marker size 50 and thick regression lines
```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="smoker",col="sex",row="time")# Faceted regression plots of total_bill vs tip by smoker, sex, and time

```

```python
sns.lmplot(x="total_bill",y="tip",data=tips,hue="smoker",col="sex",row="time",aspect=.6,palette="coolwarm")# Faceted regression plots of total_bill vs tip by smoker, sex, and time with coolwarm palette
```

```python
res=sns.kdeplot(data=tips,x=tips["total_bill"],y=tips["tip"],fill="green")# 2D KDE plot of total_bill vs tip with green fill
```

```python
res=sns.kdeplot(data=tips,x="total_bill",y="tip",fill="yellow")# 2D KDE plot of total_bill vs tip with yellow fill
```
