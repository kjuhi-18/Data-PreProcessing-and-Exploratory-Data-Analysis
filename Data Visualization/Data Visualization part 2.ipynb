```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(0,5,11)#finding 11 numbers between 0 to 5
y=x**2#square of x
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,color="r")
plt.xlabel("X axis tittle")
plt.ylabel("Y axis tittle")
plt.title("Tittle")
plt.show()
#lineplot of x and y
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"r--")#plotting x,y,of red color line as -- line
plt.subplot(1,2,2)
plt.plot(y,x,"g*-")#plotting y,x of green color 
```

```python
fig=plt.figure()#creating blank canvas
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,"b")
#seting xlabels,ylabels,title
axes.set_xlabel("X axis tittle")
axes.set_ylabel("Y axis tittle")
axes.set_title("Tittle")
```

```python
fig=plt.figure()#creating blank canvas
axes1=fig.add_axes([0.1,0.1,0.8,0.8])#main axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3])#inset axes
#larger plot
axes1.plot(x,y,"b")
#seting xlabels,ylabels,title
axes1.set_xlabel("X axis_1 tittle")
axes1.set_ylabel("Y axis_1 tittle")
axes1.set_title("axes 1Tittle")
#inset plot

axes2.plot(y,x,"k")
#seting xlabels,ylabels,title
axes2.set_xlabel("X axis_2 tittle")
axes2.set_ylabel("Y axis_2 tittle")
axes2.set_title("axes 2 Tittle")
```

```python
fig,axes=plt.subplots()#empty canvases
axes.plot(x,y,"r")
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("title")


```

```python
#empty canvas of 1 by 2 subplots
fig,axes=plt.subplots(nrows=1,ncols=2)
```

```python
#axes is an array of axes to plot on
axes
```

```python
#iterating thorugh axes
for ax in axes:
    ax.plot(x,y,"b")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.set_title("title")
fig
```

Styling line colour and style


```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots()

ax.plot(x,x**2,"b.-")#line colour blue and .- pattern
ax.plot(x,x**3,"g--")#line colour black and -- pattern
```

```python
fig,ax=plt.subplots()
ax.plot(x,x+1,"b",alpha=0.5)
ax.plot(x,x+2,color="#8B008B")#changing the colours
ax.plot(x,x+3,color="#FF8C00")


```

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.plot(x,x+1,color="red",linewidth=0.25)
ax.plot(x,x+2,color="red",linewidth=0.5)
ax.plot(x,x+3,color="red",linewidth=1)
ax.plot(x,x+4,color="red",linewidth=2)
#possible line styles
ax.plot(x,x+5,color="green",lw=3,linestyle="-")
ax.plot(x,x+6,color="green",lw=3,linestyle="-.")
ax.plot(x,x+7,color="green",lw=3,linestyle=":")






```
