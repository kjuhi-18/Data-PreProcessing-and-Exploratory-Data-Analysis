```python
import matplotlib.pyplot as plt
data=[[30,25,50,20],[40,23,51,17],[35,22,45,19]]
X=np.arange(4)
plt.bar(X+0.00,data[0],color="b",width=0.25)#multiple bars in a single plot
plt.bar(X+0.25,data[1],color="g",width=0.25)
plt.bar(X+0.50,data[2],color="r",width=0.25)



```

```python
N=5
menMeanScore=(20,35,30,35,27)
womenMeanScore=(25,32,34,20,25)
ind=np.arange(N)
width=0.35
plt.bar(ind,menMeanScore,width,color="blue")#plotting the men score
plt.bar(ind,womenMeanScore,width,bottom=menMeanScore,color="k")#plotting the women'score above the men
plt.ylabel("Scores")
plt.title("Scores by group and gender")
plt.xticks(ind,("G1","G2","G3","G4","G5"))
plt.yticks(np.arange(0,81,10))
plt.legend(labels=["MEN","WOMEN"])
plt.show()
```
