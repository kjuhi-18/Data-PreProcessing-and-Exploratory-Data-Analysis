```python
import pandas as pd
import numpy as np
```

1.Min_max scaler


import Min_max scaler(normalization_scaler)

```python
from sklearn.preprocessing import MinMaxScaler#importing min max scaller from the sklearn
```

Define Data

```python
data=pd.DataFrame(data=[[5,5],[10,12],[2,10],[0,10]],columns=['X','Y'])#making a sample dataframe
```

```python
import matplotlib.pyplot as plt
plt.scatter(data.X,data.Y)
plt.axvline(x=0,c='red')
plt.axhline(y=0,c='yellow')
plt.show()
# Scatter plot of DataFrame points with X=0 (red) and Y=0 (yellow) reference lines
```

Define FIT and transform min-max-scaler


```python
scaler=MinMaxScaler()
min_max=scaler.fit_transform(data)
```

```python
min_max#the values after scalling
```

Print Scaled data

```python
min_max=pd.DataFrame(min_max,columns=["X","Y"])#converting back to dataframe 
```

```python
min_max#the new dataframe
```

PLOT BOTH MIN_MAX SCALER

```python
import matplotlib.pyplot as plt
plt.scatter(min_max.X,min_max.Y)
plt.axvline(x=0,c='red')
plt.axhline(y=0,c='yellow')
plt.show()
#plotting the min max dataframe's both columns
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(data.X,data.Y)
plt.axvline(x=0,c='red')
plt.axhline(y=0,c='yellow')

plt.subplot(1,2,2)
plt.scatter(min_max.X,min_max.Y)
plt.show()
#plotting both dataframes to see the change
```

2.Standard Scaler

import standard scaler(standardization_scaler)

```python
from sklearn.preprocessing import StandardScaler#importing standard scaler from the sklearn
```

```python
data=pd.DataFrame(data=[[5,5],[10,2],[2,0],[0,10]],columns=["X","Y"])
data#again creating the dataframe
```

```python
scaler1=StandardScaler()
std_scal=pd.DataFrame(scaler1.fit_transform(data),columns=["X","Y"])#transforming the dataframe

std_scal
```

```python
std_scal.describe()
```

Plot standard scaler

```python
import matplotlib.pyplot as plt
plt.scatter(std_scal.X,std_scal.Y)
plt.axvline(x=0,c='red')
plt.axhline(y=0,c='yellow')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(data.X,data.Y)
plt.axvline(x=0,c='red')
plt.axhline(y=0,c='yellow')

plt.subplot(1,2,2)
plt.scatter(std_scal.X,std_scal.Y)
plt.show()
#plotting both the dataframes to see the changes
```

ploting both the scaler

```python
import matplotlib.pyplot as plt
plt.scatter(std_scal.X,std_scal.Y,marker="^")
plt.scatter(min_max.X,min_max.Y,marker="o")

plt.axvline(x=0,c='red')
plt.axhline(y=0,c='yellow')
plt.show()
```
