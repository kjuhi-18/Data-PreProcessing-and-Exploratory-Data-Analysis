Scrap sit website

```python
from bs4 import BeautifulSoup
```

```python
import requests
import os
```

```python
url="https://www.sitpune.edu.in/"
```

```python
response=requests.get(url)
```

```python
print(response)
```

```python
html_content=response.content
soup=BeautifulSoup(html_content,'html.parser')
```

Extracting the tittle

```python
soup.find("title")
```

```python
soup.find("title").string
```

Extrcting h1 tags

```python
h1_tags=soup.find_all('h1')
```

```python
h1_tags
```

```python
for h1 in h1_tags:
    print(h1.text)
```

h2 tags:


```python
h2_tags=soup.find_all('h2')
for h2 in h2_tags:
    print(h2.text)
```

h3 tags

```python
h3_tags=soup.find_all('h3')
for h3 in h3_tags:
    print(h3.text)
```

Extracting the a tags

```python
a_tags=soup.find_all('a')
for a in a_tags:
    print(a.text)
```

```python
len(a_tags)
```

Extracting the links from the a tags

```python
list_of_links=[]
for a in a_tags:
    link=a.get('href')
    if link !=None:
        list_of_links.append(link)
```

```python
len(list_of_links)
```

```python
list_of_links
```

Making the csv file

```python
import pandas as pd
```

makinga df of the links

```python
df_links=pd.DataFrame(data=list_of_links,columns=['Link'])
```

Extracting all the images from web

```python
img_links=[]
link=soup.select('img[src^="https://www.sitpune.edu.in/assets/images"]')
```

```python
for img in link:
    img_links.append(img['src'])
```

```python
directory = r"C:\Users\Kunal jhindal\OneDrive\Desktop\Sem3\DPEDAL\scrap_images"

os.makedirs(directory, exist_ok=True)
```

```python
for index,img_link in enumerate(img_links):
    img_data=requests.get(img_link).content
    with open(f'{directory}/'+str(index+1)+".jpg","wb+") as f:
        f.write(img_data)
```
